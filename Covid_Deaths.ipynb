```python
import pandas
import csv
import os
import matplotlib.pyplot as plt
import json

file_to_load = os.path.join("downloads", "covid_deaths_data.csv")

covid_deaths = pd.read_csv(file_to_load)
covid_deaths.head()
```

```python
query = 'California'
dm = covid_deaths[covid_deaths['State'] == query]
dm.head(63)
```

```python
Covid_deaths = dm['COVID-19 Deaths'].sum() /10
Pneumonia_deaths = dm['Pneumonia Deaths'].sum() /10
Influenza_deaths = dm['Influenza Deaths'].sum() /10 
PC_deaths = dm['Pneumonia and COVID-19 Deaths'].sum() /10 
PCI_deaths = dm['Pneumonia, Influenza, or COVID-19 Deaths'].sum() /10
print(Covid_deaths)
print(Pneumonia_deaths)
print(Influenza_deaths)
print(PC_deaths)
print(PCI_deaths)
```

```python
summary_df = pd.DataFrame({"Covid Deaths  " : [Covid_deaths],
                            "Pneumonia Deaths ": [Pneumonia_deaths],
                            "Influenza Deaths  " : [Influenza_deaths],
                          "Pneumonia and Covid" : [PC_deaths],
                          "Pneumonia, Covid, or Influenza": [PCI_deaths]})
summary_df
```

```python
ax = summary_df.plot(kind='bar', title = "Death Totals in Calfornia", figsize =(15,10), legend=True, fontsize=12)
ax.set_xlabel("Covid Related Diseases")
ax.set_ylabel("Number of Deaths")
```

```python
LA_file = os.path.join("downloads", "Provisional_COVID-19_Death_Counts_in_the_United_States_by_County.csv")

LA_data = pd.read_csv(LA_file)
LA_data.head()
```

```python
query = 'Los Angeles County'
query2 = 'Riverside County'
query3 = 'Ventura County'
query4 = 'Orange County'
query5 = 'San Bernardino County'
dsLA = LA_data[LA_data['County name'] == query]
dsR = LA_data[LA_data['County name'] == query2]
dsV = LA_data[LA_data['County name'] == query3]
dsO = LA_data[LA_data['County name'] == query4]
dsSB = LA_data[LA_data['County name'] == query5]

C1 = dsLA['Deaths involving COVID-19'].sum()
C2 = dsR['Deaths involving COVID-19'].sum()
C3 = dsV['Deaths involving COVID-19'].sum()
C4 = dsO['Deaths involving COVID-19'].sum()
C5 = dsSB['Deaths involving COVID-19'].sum()

summary_LA = pd.DataFrame({"Los Angeles County " : [C1],
                            "Riverside County": [C2],
                            "Ventura County  " : [C3],
                          "Orange County" : [C4],
                          "San Bernadino County": [C5]})
summary_LA
```

```python
query = 'CA'
CA = LA_data[LA_data['State'] == query]
CA.head()
```

```python
CA_plot = summary_LA.plot(kind='bar', title ="Covid Deaths in Greater Los Angeles", figsize=(15, 10), legend=True, fontsize=12)
CA_plot.set_ylabel("Deaths")
CA_plot.set_xlabel("LA                         Riverside               Ventura                  Orange         San Bernadino")

```
